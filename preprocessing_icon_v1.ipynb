```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import PIL
import h5py
```

## LLD-icon

```python
# LLD-icon- path
path = "icon/"

# simple way to load the complete dataset (for a more sophisticated generator example, see LLD-logo script)
# open hdf5 file
hdf5_file = h5py.File(path + 'LLD-icon.hdf5', 'r')

# load data into memory as numpy array
images, labels = (hdf5_file['data'][:], hdf5_file['labels/resnet/rc_64'][:])

# alternatively, h5py objects can be used like numpy arrays without loading the whole dataset into memory:
images, labels = (hdf5_file['data'], hdf5_file['labels/resnet/rc_64'])
```

```python
X = np.transpose(images[0], axes=[1, 2, 0])
X = np.uint8(X)  

plt.imshow(X)
plt.axis('off')
plt.title(labels[0], fontsize=16)
plt.show()
```

```python
fig, axs = plt.subplots(3, 3, figsize=(18, 10))
fig.subplots_adjust(hspace = .5, wspace=.001)

axs = axs.ravel()

for i in range(9):

    axs[i].imshow(np.transpose(images[i], axes=[1, 2, 0]))
    axs[i].set_title(labels[i], fontsize = 22)
    axs[i].axis('off')
    
fig.savefig('icon.jpg')
```
